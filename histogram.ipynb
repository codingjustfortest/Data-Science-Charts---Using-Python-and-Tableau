```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('SuperStore_data.csv')
plt.figure(figsize=(4,2), dpi=100)
sns.histplot(np.log1p(df['Sales']), bins=50, color='blue', kde=True)
plt.title('Sales Dist.')
plt.ylabel('Frequency')
plt.show()
```

```python
regions = df['Region'].unique()

for region in regions: # [0:2]:
    sns.histplot(np.log1p(df[df['Region'] == region]['Sales']), bins=30, alpha=0.5, label=region, kde=True)
plt.legend(title='Region')
```
