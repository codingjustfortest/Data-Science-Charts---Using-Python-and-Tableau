```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(dpi=100)
plt.rcParams['figure.figsize'] = (5, 5)
ax = fig.add_subplot(111, projection="3d")

xx = np.arange(-20,20,0.1)
yy = np.arange(-20,20,0.1)
X, Y = np.meshgrid(xx , yy)
Z = X ** 2 + Y ** 2

ax.plot_surface(X, Y, Z, cmap="viridis", lw=0.5, rstride=1, cstride=1)
ax.contour(X, Y, Z, 10, lw=2, cmap="viridis", linestyles="solid", offset=-1)
ax.contour(X, Y, Z, 10, lw=2, colors="k", linestyles="solid")
plt.show()
```
