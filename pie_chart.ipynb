```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('SuperStore_data.csv')
# df.shape
# df.head()
# df.info()
# df['Category']

sales_by_region = df.groupby('Region')['Sales'].sum()
sales_by_segment = df.groupby('Segment')['Sales'].sum()

print(sales_by_region.index)
print(sales_by_region.values)

plt.figure(figsize=(4,4), dpi=200)
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%')
plt.title('Segment By Region')
plt.show()
```

```python
plt.figure(figsize=(4,4), dpi=200)
plt.pie(sales_by_segment, labels=sales_by_segment.index, autopct='%1.1f%%')
plt.title('Segment By Segment')
plt.show()
```
