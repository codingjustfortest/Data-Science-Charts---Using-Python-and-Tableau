```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('SuperStore_data.csv')

# Convert 'Order Date' to datetime, handle errors
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df.info()
# Print rows with NaT values in 'Order Date' to inspect
# print(df[df['Order Date'].isnull()])
df = df.set_index('Order Date')
monthly_df = df.resample('ME').sum()
```

```python
# Sale PLot

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4)) 

ax[0].plot(monthly_df.index, monthly_df['Sales'], color='r')
ax[0].set_title('Monthly Sales')
ax[0].set_xlabel('Date')
ax[0].set_ylabel('Sales')
ax[0].tick_params(axis='x', rotation=45)

ax[1].plot(monthly_df.index, monthly_df['Profit'], color='g')
ax[1].set_title('Monthly Profit')
ax[1].set_xlabel('Date')
ax[1].set_ylabel('Profit')
ax[1].tick_params(axis='x', rotation=45)
```
